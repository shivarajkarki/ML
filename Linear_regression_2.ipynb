```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = []
Y = []

for line in open('machine_learning_examples-master/linear_regression_class/data_2d.csv'):
    x1, x2, y = line.split(',')
    X.append([1, float(x1), float(x2)])
    Y.append(float(y))

X = np.array(X)
Y = np.array(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,0], X[:,1], Y)
ax.set_xlabel('X0')
ax.set_ylabel('X1')
ax.set_zlabel('y')
plt.show
```

Calculate weights

```python
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat = np.dot(X, w)
```

compute r-squared

```python
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1- (d1.dot(d1)/d2.dot(d2))
print(r2)
```
